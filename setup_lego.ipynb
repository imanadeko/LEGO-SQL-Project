```python
# Import necassary libraries
import pandas as pd
import os
import urllib
from sqlalchemy import create_engine, text
```

```python
# Configuration for database connection
server = r'iman\SQLEXPRESS'
database = 'lego'
folder_path = r'C:\Users\USER\Desktop\Data Projects\LEGO SQL Project\datasets'
```

```python
# Connect to the database to upload csv files
params = urllib.parse.quote_plus(
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    f'SERVER={server};'
    f'DATABASE={database};'
    r'Trusted_Connection=yes;'
)
lego_engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
# Loop through all csv files in the folder and upload them to the database
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        table_name = file.replace('.csv', '')
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        df.to_sql(table_name, lego_engine, if_exists='replace', index=False)
        print(f'Uploaded {file} to table {table_name}')
```
